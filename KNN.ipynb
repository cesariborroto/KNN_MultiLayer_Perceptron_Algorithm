In this exercise, Train the MultiLayer Perceptron Algorithm for Wine Dataset.

```python
#Cesar Borroto
#SID 6041526
import pandas as pd
import numpy as np
import matplotlib as plt
# Download the dataset from https://archive.ics.uci.edu/ml/datasets/wine
# Load the wine dataset
```

```python
wine = pd.read_csv('wine.data', names = ["Cultivator", "Alcohol", "Malic_acid", "Ash", "Alcalinity_of_ash", "Magnesium", "Total_phenols", "Flavanoids", "Nonflavanoid_phenols", "Proanthocyanins", "Color_intensity", "Hue", "OD280", "Proline"])
```

```python
# Print the top 5 rows
wine.head()
```

```python
# Check the statistical values of each features
wine.describe()
```

```python
# Print the shape of data 
wine.shape
```

```python
# Divide data into features and target column as X and y respectively
X = wine.drop("Cultivator",axis=1)
y = wine["Cultivator"]
```

```python
X.head()
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
# Split the data into train and test
X_train.shape
```

```python
# Scale the data using standardScaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=500)
```

```python
mlp.fit(X_train,y_train)
```

```python
# Develop a MLP architecture as described in class and train the model.
predictions=mlp.predict(X_test)
```

```python
# Perform the predictions and generate the classification report and confusion matrix
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
